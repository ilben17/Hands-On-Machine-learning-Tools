<p> Seaborn is the best library for Data Visualization. Seborn allows you to create more beautiful and more worked graphics than maplotlib, by writing only one line of code! </p>
<p> Seaborn is perfect for viewing Dataset Pandas, and allows you to do a good statistical exploration of the different variables in your data. It is built on the basis of pandas and matplotlib </p>

<p> The pairplot function of searborn is ideal for having a good overview of your dataset. Pairplot allows you to see the relationship between each variable, and also produces histograms, perfect for seeing how your data is distributed according to the different variables. </p>

<p> Seaborn also allows you to view the data that forms categories thanks to seaborn boxplot. Boxplots are ideal and very powerful </p>

<p> Seaborn is very easy to use, there are few functions, and they almost all follow the same structure: (x, y, data, hue) </p>

catplot (x, y, data, hue) </br>
boxplot (x, y, data, hue) </br>
scatterplot (x, y, data, hue) </br>
etc ... </br>

<p> Additional functions to explore: Seaborn distplot () jointplot () and heatmap () which are widely used in Data Science and Data Mining practice. </p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## pairplot function

```python
iris = pd.read_csv('iris.csv')
iris.head() # we have 5 variables
```

```python
sns.pairplot(iris, hue = 'species')
```

## Categorical plots: 
<p>e.g: Dans titanic il existe bcp de categories (survived avec value 1 ou 0, pclass avec value 1,2 ou 3). Les fonction affichees avec pairplot seront en forme de ligne vericaux ou horizontaux</p>

<p style="font-weight:bold;">In this case, it's preferable to use as example the methodes like: catplot() or boxplot()</p>
<p> (Always same way to write these functions:  <span style="color:blue;font-weight:bold;">sns.method(x, y, data, hue, size, style)</span>)</p>

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

## Pour viualiser la distribution de la variable 'tarif'

```python
sns.distplot(titanic['fare'])
```

## Visualiser des matrice avec heatmap()

```python
sns.heatmap(titanic.corr())
```
