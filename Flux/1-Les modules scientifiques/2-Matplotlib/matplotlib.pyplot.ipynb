```python
##################### Pyplot ##################

import matplotlib.pyplot as plt
import numpy as np

###################### Avec la function PLOT() ##########################

X = np.linspace(start=0, stop=2, num=6)
X # array([0. , 0.4, 0.8, 1.2, 1.6, 2. ])

Y = X ** 2 # rappel: Brodcasting va s'opperer ici
Y # array([0. , 0.16, 0.64, .... ])

plt.plot(X, Y) # <---------Important: faut que le # de donnees dans X = # de donnnees dans Y

# si on utilise autre chose que jupiter il rajout la ligne (plt.show()) <----IMPORTANTE

###################### Avec la function sccatter() ##########################
plt.scatter(X, Y)

###################### # customisr une courbe ##################
plt.plot(X, Y, c="red", lw='3', ls='--')
```

```python
#The right way to draw figure (cycle de vie d'une figure)

#debut:plan
plt.figure(figsize=(12,8))

#contenu
plt.plot(X,Y, label='quadratic')
plt.plot(X, X**3, label='cubic')
plt.title("Figure 1")
plt.xlabel('axe x')
plt.ylabel("axe y")
plt.legend()

#display
plt.show()

plt.savefig('testfig.png')
```

```python
################################### AVEC OOP (non trop utilisee mais plus de possibilties) ################################

fig, axes = plt.subplots() # <---- subplots Avec un s

axes.plot(X, Y)
plt.show() #<---- ca donne la meme chose

type(fig) # Object matplotlib.figure.Figure
type(axes) # matplotlib.axes._subplots.AxesSubplot

############## Mainteant si on veut creer pusieurs graphiques et les faire partager le meme axe X par ex ##########

fig2, axes2 = plt.subplots(2, 1, sharex=True)

type(fig2) # matplotlib.figure.Figure
type(axes2) # BUT: numpy.ndarray <--------- DIFFERENCE is in subplots() without args we had axes as AxesSubplot Object  

axes2 # is array of AxesSubplot [AxesSubplot object, AxesSubplot object] with axes2.shape = (2,) i.e (1,2)
        
axes2[0].plot(X, Y) # <--- pour accceder au plan de travail 1
axes2[1].plot(X, np.sin(X)) # <--- pour accceder au plan de travail 2

plt.show()

# ---> Important si par exmple j'ai ecris:  fig3, axes3 = plt.subplots(2, 2, sharex=True)
#      => axes3 = [[AxesSubplot, AxesSubplot],[AxesSubplot, AxesSubplot]] i.e axes3.shape = (2,2)
#      => axes3[0][0].plot(X, X**3)
#      => axes3[0][1].plot(X, X**5)
#      => axes3[1][0].plot(X, np.cos(X))
#      => axes3[1][1].plot(X, np.exp(X))

fig3, axes3 = plt.subplots(2, 2, sharex=True)
axes3[0][0].plot(X, X**3)
axes3[0][1].plot(X, X**5)
axes3[1][0].plot(X, np.cos(X))
axes3[1][1].plot(X, np.exp(X))

plt.show()
```

```python
# Exercice :
np.random.seed(10)

dict_experiences = {f"experience{i}" : np.random.randn(4, 2) for i in range(3)} 

dict_experiences
#{'experience0': [[ 1.3315865 ,  0.71527897],[-1.54540029, -0.00838385],[ 0.62133597, -0.72008556],[ 0.26551159,  0.10854853]],
# 'experience1': [[ 0.00429143, -0.17460021],[ 0.43302619,  1.20303737],[-0.96506567,  1.02827408],[ 0.22863013,  0.44513761]],
# 'experience2': [[-1.13660221,  0.13513688],[ 1.484537  , -1.07980489],[-1.97772828, -1.7433723 ],[ 0.26607016,  2.38496733]]}

#une colonne represente 1 variable so in our dic we have 2 variables eg: 
# pairs (x,y) de la variable 1 in experience0 : (x=i=0, y=1.3315865), (1, -1.54540029), (2, 0.62133597) ...
# pairs (x,y) de la variable 2 in experience0 : (x=i=0, y=0.71527897), (1, -0.00838385), (2, -0.72008556) ...

def graphiques_experiences(dataset): # afficher un garpahique par experience        
    number_rows = len(dataset)
    plt.figure(figsize=(12, 15))
    for key, position in zip(dataset.keys(), range(1, number_rows + 1)):  #rmq: si range(number_rows ) --> pb car position dans subplot() ne peut pas etre = 0
        plt.subplot(3, 1, position)
        plt.plot(dataset[key])
        plt.title(key)
    plt.show()

graphiques_experiences(dict_experiences)    
```

```python
graphiques_experiences({'experience0': [10, 10, 10, 10],'experience1': [10, 5, 10, 5], 'experience2': [1, 2, 3, 1]})
```
