```python

# importons Iris flower dataset (voir wikepedia)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

x = iris.data
y = iris.target # class 0 , 1 ou 2
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables') # x contient 150 exemples et 4 variables. 
                                                                     # 150 c<est le nombre de lignes
                                                                     # une colonne correspond a une variable et on a tableau de (150, 4)
print(f'Il y a {np.unique(y).size} classes') # Il y a 3 classes 

#x
# [[5.1, 3.5, 1.4, 0.2],
# [4.9, 3. , 1.4, 0.2],
# [4.7, 3.2, 1.3, 0.2],
# [4.6, 3.1, 1.5, 0.2],
# .....
```

```python
# Position 5 : GRAPHIQUE DE CLASSIFICATION WITH plt.scatter. Limite: on ne peut representer que deux variables

plt.scatter(x[:, 0], x[:, 1], c=y, alpha = 0.7, s=x[:, 2]*100) 
# c=y i.e color = 0, 1 or 2 se qui va se refleter par des couleurs differentes sur le graphic
#alpha : ppour controler la transparence
#s : pour controler la taille du point. s=x[:, 2]*100: multiplier la variable 2 par 100 
                                        # => plus la taille du variable est grande plus le petal est long 

plt.xlabel('Longueur Sepal')
plt.ylabel('Largeur Sepal')
```

```python
# Position 4 : GRAPHIQUE 3D (pour visualiser plus de 2 variables sur un meme graphique)
#e.g 1:
#from mpl_toolkits.mplot3d import Axes3D # ss-package de matplotlib

#ax = plt.axes(projection='3d')
#ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y) #<--------- Attention: S'ouvre in separate onglet

#e.g2: visualiser une fonction mathematique
f= lambda x, y : np.sin(x) + np.cos(x + y)

X = np.linspace(0, 5, 5) # values between 0 and 5. 100 items X: [0.  , 1.25, 2.5 , 3.75, 5.  ]
Y = np.linspace(0, 5, 5) # Y: [0.  , 1.25, 2.5 , 3.75, 5.  ]

X, Y = np.meshgrid(X, Y) #(voir doc word assi pour mshgrid())

X
# [[0.  , 1.25, 2.5 , 3.75, 5.  ],
# [0.  , 1.25, 2.5 , 3.75, 5.  ],
# [0.  , 1.25, 2.5 , 3.75, 5.  ],
# [0.  , 1.25, 2.5 , 3.75, 5.  ],
# [0.  , 1.25, 2.5 , 3.75, 5.  ]]

Y
# [[0.  , 0.  , 0.  , 0.  , 0.  ],
# [1.25, 1.25, 1.25, 1.25, 1.25],
# [2.5 , 2.5 , 2.5 , 2.5 , 2.5 ],
# [3.75, 3.75, 3.75, 3.75, 3.75],
# [5.  , 5.  , 5.  , 5.  , 5.  ]]

Z = f(X, Y)
Z.shape #(5,5)
# [[ 1.        ,  1.26430698, -0.20267147, -1.39212068, -0.67526209],
# [ 0.31532236,  0.147841  , -0.22208721, -0.28789913,  0.04052514],
# [-0.80114362,  0.12842526,  0.88213433,  0.4278881 , -0.61228896],
# [-0.82055936,  1.2326468 ,  1.59792156, -0.224926  , -1.73976996],
# [ 0.28366219,  1.94843404,  0.94510746, -1.352407  , -1.7979958 ]]

#ax.plot_surface(X, Y, Z, cmap= "plasma")
```
